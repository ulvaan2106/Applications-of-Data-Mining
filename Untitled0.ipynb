```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = {
    "Feature1": [10, 20, 30, 40, 50],
    "Feature2": [100, 200, 300, 400, 500],
    "Feature3": [1, 2, 3, 4, 5]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
```

```python
min_max_scaler = MinMaxScaler()
normalized_data = min_max_scaler.fit_transform(df)

print("\nMin-Max Normalized Data:")
print(pd.DataFrame(normalized_data, columns=df.columns))
```

```python
standard_scaler = StandardScaler()
standardized_data = standard_scaler.fit_transform(df)

print("\nStandardized Data:")
print(pd.DataFrame(standardized_data, columns=df.columns))
```
